I have completed the analysis of the ShopEase dataset (1,000 orders). Here are the most important things I found:

Most Important Insight: Our platform is very popular with the 46+ age group, and Beauty and Electronics are our strongest categories.

Year-on-Year Observation: The business is growing, with revenue increasing from ₹62.7 Lakhs in 2023 to ₹64.7 Lakhs in 2024.

Underperforming Area: The Clothing category is struggling with high returns (18.2%), mainly because customers "changed their mind" after ordering. Also, delivery in the Central region is slower than in other areas.

Actionable Recommendation: We should improve product descriptions in the Clothing section to reduce returns and optimize our delivery route in the Central region to increase speed.
